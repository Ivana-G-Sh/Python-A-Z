```python
#wrapper for all the code for plotting
#function is called f1
def myplot(playerlist): #myplot is name of the function
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c = 'Black', ls='--', marker = 's', ms = 8, label=name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons) 
```

```python
#we want colours to change
#define dictionary for colours
def myplot(playerlist): #myplot is name of the function
    Col = {"KobeBryant":'Black',"JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Magenta"}
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c =Col[name], ls='--', marker = 's', ms = 8, label=name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons) 
```

```python
#fix the markers
def myplot(playerlist): #myplot is name of the function
    Col = {"KobeBryant":'Black',"JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Magenta"}
    M = {"KobeBryant":'s',"JoeJohnson":"o","LeBronJames":"^","CarmeloAnthony":"D","DwightHoward":"s"}
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c =Col[name], ls='--', marker = M[name], ms = 8, label=name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons) 
```

```python
#fix the input - which matrix
def myplot(matrix, playerlist): #myplot is name of the function, playerlist je lista igraca
    Col = {"KobeBryant":'Black',"JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Magenta"}
    M = {"KobeBryant":'s',"JoeJohnson":"o","LeBronJames":"^","CarmeloAnthony":"D","DwightHoward":"s"}
    for imena in playerlist:
        plt.plot(matrix[Pdict[imena]], c =Col[imena], ls='--', marker = M[imena], ms = 8, label=imena)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons) 
```
